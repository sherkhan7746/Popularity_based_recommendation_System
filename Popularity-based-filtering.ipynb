## Popular-Based Filtering

### Load the data

```
import pandas as pd
```

```
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
credits = pd.read_csv("credits.csv")
```

```
movies.head()
```

```
credits.head()
```

### Calculate The weighted Average

```
m = movies["vote_count"].quantile(0.90)
m
```

```
C = movies["vote_average"].mean()
C
```

```
movies_filterd = movies.copy().loc[movies["vote_count"] >= m]
```

```
def weighted_rating(df, m=m, C=C):
    R = df["vote_average"]
    v = df["vote_count"]
    wr = ((v / v+m ) * R) + ((m / (v+m)) * C)
    return wr
```

```
movies_filterd["weighted_rating"] = movies_filterd.apply(weighted_rating, axis=1)
```

```
movies_filterd.sort_values("weighted_rating", ascending=False)[["title", "vote_count", "vote_average", "weighted_rating"]].head(20)
```
